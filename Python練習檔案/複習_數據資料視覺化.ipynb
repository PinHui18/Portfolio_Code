### 繪製折線圖：plot

```python
import matplotlib.pyplot as plt

listx = [1,5,7,9,13,16]
listy = [15,50,80,40,70,50]
plt.plot(listx, listy)
plt.show()
```

```python
plt.plot(listx, listy, 'g--o') # g:green, --: 虛線, o: 標記樣式
plt.show()
```

```python
plt.plot(listx, listy, color='red', lw='2.0', ls='--', label="label")
plt.legend()
```

```python
plt.title('Chart Title', fontsize=20) # 圖表標題
plt.xlabel('x-label', fontsize=14) # x座標標題
plt.ylabel('y-label', fontsize=14)
plt.plot(listx, listy) # g:green, --: 虛線, o: 標記樣式
plt.show()
```

```python
plt.xlim(0,20) # 設定x座標範圍
plt.ylim(0,100)
plt.plot(listx, listy, 'g--o') 
plt.show()
```

```python
plt.grid(color='red', linestyle=':', linewidth=1, alpha=0.5)
# 設定格線
```

```python
# 同時繪製多組資料
listx1 = [1,5,7,9,13,16]
listy1 = [15,50,80,40,70,50]
plt.plot(listx1, listy1, 'r-.s')
listx2 = [2,6,8,11,14,16]
listy2 = [10,40,30,50,80,60]
plt.plot(listx2, listy2, 'y-s')
plt.show()
```

```python
import matplotlib.pyplot as plt
year = [2015, 2016, 2017, 2018, 2019]
city1 = [128,150,199,180,150]
plt.plot(year, city1, 'r-.s', lw=2, ms=10, label='Taipei')
city2 = [120,145,180,170,120]
plt.plot(year, city2, 'g--*', lw=2, ms=10, label='Taichung')
plt.legend()
plt.ylim(50,250)
plt.xticks(year)
plt.title('Sales Report', fontsize=18)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Million', fontsize=12)
plt.grid(color='k', ls=':', lw=1, alpha=0.5)
plt.show()
```

### 繪製長條圖：bar

```python
listx = ['c', 'c++', 'c#', 'java', 'python']
listy = [45,28,38,32,50]
plt.bar(listx, listy, width=0.5, color='rgb') # 橫條圖: barh
plt.title('資訊程式課程選修人數')
plt.xlabel('程式課程')
plt.ylabel('選修人數')
plt.rcParams['font.sans-serif'] = 'Microsoft JhengHei' # 可正常顯示中文
plt.show()
```

### 繪製圓餅圖：pie

```python
size = [25,30,15,10]
labels = ['北部','西部','南部','東部']
colors = ['red','green','blue','yellow']
explode = (0,0,0.2,0)
plt.pie(size,
       explode=explode,
       colors = colors,
       labeldistance = 1.1,
       autopct = '%2.1f%%',
       pctdistance = 0.6,
       shadow = True,
       startangle = 90)  # 從90度開始繪製
```

### 在圖表區加入多張圖表

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[8,8])
plt.subplot(211)
plt.title(label='Chart 1', fontsize=20)
plt.plot([1,2,3], 'r:o')

plt.subplot(212)
plt.title(label='Chart 2', fontsize=20)
plt.plot([1,2,3], 'g--o')
plt.show()
```
