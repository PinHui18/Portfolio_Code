```python
# 靜態繪圖套件
import pandas as pd
import matplotlib.pyplot as plt

# 動態繪圖套件
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python

# read data
data = pd.read_csv('carseats2.csv')

# 看一下有什麼欄位
data.columns
```

```python

# ---------- 靜態繪圖 1 - 人口與銷售額 --------------
# 取出 x =Population   y = Sales
x = data['Population']
y = data['Sales']

# 劃出 Population Sales 點陣圖
# 設定軸大小
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(x, y )
plt.show()

# 動態 vs 靜態繪圖差異
'''
1.靜態在ppt（簡報）上好做呈現
2.動態網頁互動性高，舉例：A點是什麼？ 
使用鼠標指引就可以快速知道該點的狀況
即不用重新製靜態圖
'''
```

```python

# ---------- 動態繪圖 1 - 人口與銷售額 --------------
import plotly.express as px
fig = px.scatter(data, x="Population", 
                 y="Sales", 
                 hover_data = ['ID'], 
                 title = '人口與銷售額')

plot(fig, filename='圖1-人口與銷售額散佈圖.html')

```

```python

# ---------- 靜態繪圖 2 - 人口與銷售額（以貨架標示） --------------

# 取出各自 ShelveLoc x y 的資料
Baddata = data[data['ShelveLoc']=='Bad']
x_B = Baddata['Population']
y_B = Baddata['Sales']

Gooddata = data[data['ShelveLoc'] == 'Good']
x_G = Gooddata['Population']
y_G = Gooddata['Sales']

Meddata = data[data['ShelveLoc']=='Medium']
x_M = Meddata['Population']
y_M = Meddata['Sales']

# 劃出三個資料集的圖
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(x_B, y_B, label = 'Bad') # bad
ax.scatter(x_G,y_G, label = 'Good' ) # good
ax.scatter(x_M, y_M, label = 'Medium') # medium
plt.legend(bbox_to_anchor=(1.03, 0.8), loc=2)
plt.show()

```
